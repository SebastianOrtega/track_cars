```python
# Open the video file
cap = cv2.VideoCapture(
    "rtsp://admin:panamet0@192.168.0.208:554/cam/realmonitor?channel=1&subtype=1&unicast=true&proto=Onvif")


# Check if the video opened successfully
if not cap.isOpened():
    raise IOError("Error: Could not open video file.")

# Read the first frame of the video
ret, frame = cap.read()

# Check if the frame was successfully read
if not ret:
    raise IOError("Error: Could not read first frame of the video.")

print(frame.shape)
# Display the first frame of the video
plt.imshow(frame)
plt.show()

# Release the video capture object
cap.release()
```

```python
no_zone = np.array([[180, 480], [580, 0], [704, 0], [704, 480], [180, 480]])


# Create a blank image (white background)
height, width = 480, 704
frame = np.ones((height, width, 3), dtype=np.uint8) * 255


# Draw the polygon on the frame
cv2.polylines(frame, [no_zone], isClosed=True, color=(255, 0, 0), thickness=2)

# Convert BGR image to RGB for displaying with Matplotlib
frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

# Display the image using Matplotlib
plt.imshow(frame_rgb)
plt.title("No zone")
plt.show()
```
